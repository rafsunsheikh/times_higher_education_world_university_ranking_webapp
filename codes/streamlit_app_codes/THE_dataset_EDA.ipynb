## **Doctorate awarded to Academic Staff**

```python
# Extract the x and y values for regression
X_doctorate_awarded_to_academic_staff = df_doctorate_awarded_to_academic_staff['values'].values.reshape(-1, 1)
y_doctorate_awarded_to_academic_staff = df_doctorate_awarded_to_academic_staff['scores'].values

######################################################################### Fit a linear regression model ###############################################################################
regression_model_doctorate_awarded_to_academic_staff= LinearRegression()
regression_model_doctorate_awarded_to_academic_staff.fit(X_doctorate_awarded_to_academic_staff, y_doctorate_awarded_to_academic_staff)

# Predict y values based on the regression model
y_pred_doctorate_awarded_to_academic_staff = regression_model_doctorate_awarded_to_academic_staff.predict(X_doctorate_awarded_to_academic_staff)

# Plot the original data and regression line
plt.scatter(X_doctorate_awarded_to_academic_staff, y_pred_doctorate_awarded_to_academic_staff, label='Original Data')
plt.plot(X_doctorate_awarded_to_academic_staff, y_pred_doctorate_awarded_to_academic_staff, color='red', label='Linear Regression')
plt.xlabel('values')
plt.ylabel('scores')
plt.title('Linear Regression')
plt.legend()
plt.show()


# Plot the original data and regression line
plt.scatter(X_doctorate_awarded_to_academic_staff, y_doctorate_awarded_to_academic_staff, label='Original Data')
plt.plot(X_doctorate_awarded_to_academic_staff, y_pred_doctorate_awarded_to_academic_staff, color='red', label='Linear Regression')
plt.xlabel('values')
plt.ylabel('scores')
plt.title('Linear Regression')
plt.legend()
plt.show()

########################################### Predict Single Scores Linear Regression ####################################################################

pred_doctorate_awarded_to_academic_staff = np.array([[0.05]])
score_doctorate_awarded_to_academic_staff = regression_model_doctorate_awarded_to_academic_staff.predict(pred_doctorate_awarded_to_academic_staff)

print(score_doctorate_awarded_to_academic_staff)




######################################################################### Fit a Polynomial regression model ###############################################################################
degree = 4
poly_features_doctorate_awarded_to_academic_staff = PolynomialFeatures(degree=degree)
X_poly_doctorate_awarded_to_academic_staff = poly_features_doctorate_awarded_to_academic_staff.fit_transform(X_doctorate_awarded_to_academic_staff)

# Fit a linear regression model
poly_regression_model_doctorate_awarded_to_academic_staff= LinearRegression()
poly_regression_model_doctorate_awarded_to_academic_staff.fit(X_poly_doctorate_awarded_to_academic_staff, y_doctorate_awarded_to_academic_staff)

# Predict y values based on the regression model
y_poly_pred_doctorate_awarded_to_academic_staff = poly_regression_model_doctorate_awarded_to_academic_staff.predict(X_poly_doctorate_awarded_to_academic_staff)


# Original data
plt.scatter(X_doctorate_awarded_to_academic_staff, y_doctorate_awarded_to_academic_staff, color='blue', label='Data')

# Fitted polynomial curve
X_doctorate_awarded_to_academic_staff_range = np.linspace(min(X_doctorate_awarded_to_academic_staff), max(X_doctorate_awarded_to_academic_staff), 100)  # Create a range of x values
x_poly_doctorate_awarded_to_academic_staff_range = poly_features_doctorate_awarded_to_academic_staff.transform(X_doctorate_awarded_to_academic_staff_range.reshape(-1, 1))
y_poly_pred_doctorate_awarded_to_academic_staff_range = poly_regression_model_doctorate_awarded_to_academic_staff.predict(x_poly_doctorate_awarded_to_academic_staff_range)
plt.plot(X_doctorate_awarded_to_academic_staff_range, y_poly_pred_doctorate_awarded_to_academic_staff_range, color='red', label='Polynomial Regression')

plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

########################################### Predict Single Scores Polynomial Regression #########################################################
poly_pred_doctorate_awarded_to_academic_staff = np.array([[0.05]])  # Replace x_value with the value for which you want to predict y
new_data_poly_doctorate_awarded_to_academic_staff = poly_features_doctorate_awarded_to_academic_staff.transform(poly_pred_doctorate_awarded_to_academic_staff)
predicted_y_poly_doctorate_awarded_to_academic_staff = poly_regression_model_doctorate_awarded_to_academic_staff.predict(new_data_poly_doctorate_awarded_to_academic_staff)

print(predicted_y_poly_doctorate_awarded_to_academic_staff)


###################### Min-Max #########################
######### Min - Max Value ################
doctorate_awarded_to_academic_staff_min_value = df_doctorate_awarded_to_academic_staff['values'].min()
doctorate_awarded_to_academic_staff_max_value = df_doctorate_awarded_to_academic_staff['values'].max()

######### Min - Max Score ################
doctorate_awarded_to_academic_staff_min_score = df_doctorate_awarded_to_academic_staff['scores'].min()
doctorate_awarded_to_academic_staff_max_score = df_doctorate_awarded_to_academic_staff['scores'].max()

print (f"doctorate_awarded_to_academic_staff_min_value: {doctorate_awarded_to_academic_staff_min_value}")
print (f"doctorate_awarded_to_academic_staff_max_value: {doctorate_awarded_to_academic_staff_max_value}")
print (f"doctorate_awarded_to_academic_staff_min_score: {doctorate_awarded_to_academic_staff_min_score}")
print (f"doctorate_awarded_to_academic_staff_max_score: {doctorate_awarded_to_academic_staff_max_score}")


```
